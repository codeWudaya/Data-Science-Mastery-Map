```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Create a custom dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [25, 30, 35, 40, 45],
    'Income': [50000, 60000, 75000, 90000, 80000],
    'Education': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master']
}
```

```python
# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
print("Custom Dataset:")
print(df)
print()
```

## Apply visualization using Seaborn

# Bar Plot

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Name', y='Age',data=df)
plt.title('Income by Name')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(x='Age', y='Income', data=df)
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

# Box plot

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Education', y='Age', data=df)
plt.title('Age by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Age')
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(6, 5))
sns.violinplot(x='Education', y='Income', data=df)
plt.title('Income Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Income')
plt.show()
```

# Swarm Plot

```python
plt.figure(figsize=(6, 4))
sns.swarmplot(x='Education', y='Income', data=df)
plt.title('Income Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Income')
plt.show()
```

# Count Plot

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Education', data=df)
plt.title('Count of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```

# Joint Plot

```python
sns.jointplot(x='Age', y='Income', data=df)
plt.title('Joint Plot of Age and Income')
plt.show()
```

# Heatmap

```python
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Histogram

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], bins=5, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# KDE Plot

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(df['Income'], shade=True)
plt.title('Kernel Density Estimate of Income')
plt.xlabel('Income')
plt.ylabel('Density')
plt.show()
```

# Bar Plot with Hue

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='Name', y='Income', hue='Education', data=df)
plt.title('Income by Name and Education Level')
plt.xlabel('Name')
plt.ylabel('Income')
plt.legend(title='Education Level')
plt.show()
```
